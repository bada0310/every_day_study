```python
# 인수의 생일파티 
# 다익
# 각 노드들의 최단 시간 중에서 가장 오래 걸리는 노드의 시간을 출력하는 문제이다.
import heapq
INF = int(1e9)
def dijkstra(X,graph):
        q =[]
        dist = [INF]*(N+1)
        heapq.heappush(q,(0,X))
        dist[X] = 0
        
        while q:
            curr_val , curr_node = heapq.heappop(q)
            
            if dist[curr_node] < curr_val:
                continue 
            
            for next_val, next_node in graph[curr_node]:
                new_cost = curr_val + next_val
                
                if new_cost < dist[next_node]:
                    dist[next_node] =new_cost
                    heapq.heappush(q,(new_cost, next_node))
        return dist
    
T = int(input())
for tc in range(1,T+1):
    N, M, X  = map(int,input().split())
    
    graph = [[] for _ in range(N+1)] # 단방향그래프 
    rev_graph = [[] for _ in range(N+1)] 
    
    for _ in range(M):
        x, y, val = map(int,input().split())
        graph[x].append((val, y))
        rev_graph[y].append((val, x))
        
    res = dijkstra(X,graph)
    rev_res = dijkstra(X,rev_graph)
    max_dist = 0
    for i in range(1,N+1):
        if res[i] != INF and rev_res[i] != INF:
            round_trip = res[i] + rev_res[i]
            if round_trip > max_dist:
                max_dist = round_trip
    print(f'#{tc}', max_dist)
    
  
            
```

```python
# 최소비용경로
# 높이차가 존재하면 높이차이만큼 연료비가 더든다 
import heapq
INF = int(1e9)

def dijkstra(r,c):
    dist_grid = [[INF]*N for _ in range(N)]
    dist_grid[r][c] = 0
    q = []
    heapq.heappush(q,(dist_grid[r][c],r, c))
    
    while q:
        val, cr, cc = heapq.heappop(q)
        if val > dist_grid[cr][cc]:
            continue
        for dr, dc in [(0,1),(1,0),(0,-1),(-1,0)]:
            nr = cr + dr
            nc = cc + dc
            if nr < 0 or nr >= N or nc < 0 or nc >= N:
                continue
            cost = grid[nr][nc] -grid[cr][cc]+1 if grid[nr][nc] > grid[cr][cc] else 1
            
            if dist_grid[nr][nc] > dist_grid[cr][cc]+cost:
                dist_grid[nr][nc] = dist_grid[cr][cc]+cost
                heapq.heappush(q,(dist_grid[nr][nc], nr, nc))
    return dist_grid

T = int(input())
for tc in range(1,T+1):
    N =int(input())
    grid = [list(map(int,input().split())) for _ in range(N)]
    
    res = dijkstra(0,0)
    print(f'#{tc}', res[N-1][N-1])
```

```python
#10966_물놀이를 가자 
# 물이면 W 땅이면 L 
# 최소 이동횟수 그리고 
# 그 최종 이동 횟수들의 합
from collections import deque
dx = [0,1,0,-1]
dy = [1,0,-1,0]
def is_range(r,c):
    return 0<= r < N and 0<= c < M

T = int(input())
for tc in range(1,T+1):
    N, M = map(int,input().split())
    grid=[list(input()) for _ in range(N)]
    dist = [[-1]*M for _ in range(N)]
    q = deque()
    for i in range(N):
        for j in range(M):
            if grid[i][j] == 'W':
                q.append((i,j))
                dist[i][j] = 0
    
    while q:
        cr, cc = q.popleft()
        
        for i in range(4):
            nr = cr + dx[i]
            nc = cc + dy[i]
            
            if 0<= nr < N and 0<= nc < M and dist[nr][nc] == -1:
                dist[nr][nc] = dist[cr][cc] + 1
                q.append((nr,nc))
    total_dist = 0
    for i in range(N):
        for j in range(M):
            if grid[i][j] == 'L':
                total_dist += dist[i][j]
    print(f'#{tc}',total_dist)
```

```python
# MST 
# prim 
import heapq
INF = int(1e9)

def prim(x):
    q = [(0,x)]
    D[x] = 0
    ans = 0
    while q:
        val , u = heapq.heappop(q)
        if visited[u]:
            continue
        visited[u] = 1
        ans += D[u]
        
        for w, v in graph[u]:
            if not visited[v] and D[v] > w:
                D[v] = w
                heapq.heappush(q, (D[v], v))
    return ans

T =int(input())
for tc in range(1,T+1):
    N, E = map(int,input().split())
    graph = [[] for _ in range(N+1)]
    D = [INF]*(N+1) # 가중치 저장하는 리스트
    visited = [0]*(N+1) # 트리에 포함된 정점인지를 check
    for _ in range(E):
        u, v, val = map(int,input().split())
        graph[u].append((val, v))
        graph[v].append((val, u))
    res = prim(0)
    print(f'#{tc}', res)
```
