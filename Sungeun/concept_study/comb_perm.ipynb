```python
# N&M 1 (permutation)
N, M = map(int,input().split())
arr = [i for i in range(1,N+1)]
# 순열(dfs)
visited = [False]*N
result = []

def dfs(depth): # depth 현재 뽑은 갯수 
    
    if depth == M: # 길이가 같아지면 
        print(*result)
        return  
    
    for i in range(N):
        if not visited[i]:
            visited[i] = True
            result.append(arr[i])
            
            dfs(depth+ 1)
            result.pop()
            visited[i] =False

dfs(0)
```

```python
# N&M 2 (combination) 

N, M = map(int,input().split())
arr = [i for i in range(1,N+1)]

visited = [False]*N
result = []

def comb(depth, start): # dfs? 
    if depth == M:
        print(*result)
        return
    
    for i in range(start,N):
        if not visited[i]:
            visited[i] = True
            result.append(arr[i])
            
            comb(depth+1,i+1)
            result.pop()
            visited[i] = False
            
comb(0,0)
```

```python
# N & M 4
# 중복조합 (rep-combination)
N, M = map(int,input().split())
arr = [i for i in range(1,N+1)]

visited = [False]*N
result = []

def comb_rep(depth, start):
    if depth == M:
        print(*result)
        return
    
    for i in range(start,N):
        visited[i] = True
        result.append(arr[i])
        
        comb_rep(depth+1,i)
        visited[i] = False
        result.pop()
        
comb_rep(0,0)
        
```

```python
# N&M 9 
#순열 (arr 에 중복 원소가 있는)
N, M = map(int,input().split())
arr = list(map(int,input().split()))
arr.sort()

visited = [False]*N
result = []

def perm(depth):
    if depth == M:
        print(*result)
        return
    
    prev = -1  # flag
    
    for i in range(N):
        if not visited[i] and prev != arr[i]:
            visited[i] = True
            result.append(arr[i])
            prev = arr[i]
            
            perm(depth+1)
            visited[i] = False
            result.pop() 
perm(0)
```
